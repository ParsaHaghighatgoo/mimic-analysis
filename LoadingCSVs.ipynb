```python
import pandas as pd
import os

chunk_size = 30_000_000
output_dir = "E:\Chartevents\\"
# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Read and save the CSV file in chunks
chunk_iterator = pd.read_csv(path + 'chartevents.csv', chunksize=chunk_size)

# Process and save each chunk
for i, chunk in enumerate(chunk_iterator):
    output_file = f"{output_dir}chartevents_chunk_{i + 1}.csv"
    chunk.to_csv(output_file, index=False)
    print(f"Chunk {i + 1} saved to {output_file}")
```
